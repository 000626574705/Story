```python
#for drawing 3d scatter data
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)

fig =plt.figure()
ax=fig.add_subplot(projection = '3d')
ax.scatter(x,y,z)
ax.show()
```
