Q1solution-Matplotlib is a popular and widely used Python library for creating static, interactive, and animated visualizations of data. It provides a high-level interface for creating a variety of plots, charts, and graphs, making it an essential tool for data analysts, scientists, and engineers working with Python.

With Matplotlib, you can create line plots, scatter plots, bar charts, histograms, 3D plots, and many other types of visualizations to help you explore and understand your data.

```python
# Q2 solution- Scatter Plot: A plot that represents individual data points as dots on a 2D plane, commonly used to visualize the relationship between two variables.
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("this is my x axis")
plt.ylabel("this is my y label")
plt.title("this is my title")
plt.scatter(x,y)
```

```python
#Q3 solution-The subplot() function in Matplotlib's Pyplot module is used to create multiple plots or subplots within a single figure. It is particularly helpful when you want to visualize different aspects of your data side by side or compare multiple datasets simultaneously. Instead of creating separate figures for each plot, you can arrange them in a grid-like structure using subplot().
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

y2 = np.array([50, 20, 40, 20, 60, 70])

y3= np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])


plt.subplot(2, 1, 1)
plt.plot(x, y1)

```

```python
#Q4 Bar plot: A chart that represents categorical data with rectangular bars, where the length of each bar is proportional to the value it represents.

#uses of bar plot:-
 # 1.Comparison of Categories
 #2.Visualizing Frequencies
 #3.Showing Patterns and Trends
 #4.Comparison of Subgroups
 #5.Communicating Findings
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
#for vertical bar plot
plt.bar(company,profit)
```

```python
#for horizontal bar plot
plt.barh(company,profit)

```

```python
#Q5 solution- A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of numerical data based on five summary statistics. It provides a concise way to visualize the central tendency, spread, and potential outliers in the data set.
# uses of box plot-
#1.Visualizing Data Distribution:
#2.Comparing Data Groups
#3.Detecting Skewness
#4.Handling Missing Data

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box1)
```
