```python
#In Flask, the url_for() function is used for URL building. It is a built-in function provided by Flask that helps generate URLs for a given endpoint and any associated parameters.

#The url_for() function takes the name of the endpoint as the first argument and can optionally accept keyword arguments representing the parameters of the URL.

#Here's an example to demonstrate the usage of the url_for() function in Flask:
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to the home page!'

@app.route('/user/<username>')
def user_profile(username):
    return f'Hello, {username}!'

@app.route('/product/<int:product_id>')
def product_details(product_id):
    return f'Product ID: {product_id}'

with app.test_request_context():
    # Generating URLs using url_for()
    home_url = url_for('home')
    user_profile_url = url_for('user_profile', username='john_doe')
    product_details_url = url_for('product_details', product_id=123)

    # Printing the generated URLs
    print(home_url)
    print(user_profile_url)
    print(product_details_url)

```
